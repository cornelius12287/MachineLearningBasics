```python
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import pandas as pd
```

```python
admissionData = np.genfromtxt('Admitted_Rejected.csv',delimiter=',',skip_header=1)
```

```python
decision = []
for i in admissionData[:,2]:
    if (i==1.0):
        decision.append('accepted')
    else:
        decision.append('rejected')
decision = np.array(decision)
```

```python
scatter_x = admissionData[:,0]
scatter_y = admissionData[:,1]
group = decision
cdict = {'rejected':'red','accepted':'green'}
fig, ax = plt.subplots()
for g in np.unique(group):
    ix = np.where(group==g)
    ax.scatter(scatter_x[ix],scatter_y[ix],c=cdict[g],label=g)
plt.legend(labels=['Accepted','Rejected'])
plt.xlabel('Exam 1')
plt.ylabel('Exam 2')
```

```python
df = pd.DataFrame(data=admissionData,columns=['Exam1','Exam2','Decision'])
```

```python
sns.scatterplot(x='Exam1',y='Exam2',hue='Decision',data=df)
```
